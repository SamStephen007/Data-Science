# Model Building

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = df.drop(['Final Year Result','Extracurricular Activities','Distance (km)'], axis=1)
y = df['Final Year Result']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x['Internet (Yes/No)'] = le.fit_transform(x['Internet (Yes/No)'])
y = le.fit_transform(y)
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)
```

```python
input = pd.DataFrame([[8,700000,10,77,1],[8.3,300000,15,87,0]], columns=[
    'High school GPA', 'Parental Income (₹)', 'Study Hours/Week',
    'Entrance Test Score', 'Internet (Yes/No)'
])

r  = model.predict(input)
for res in r:
  if(res==1):
    print("Fail")
  elif(res==0):
    print("Distinction")
  else:
    print("Pass")
```
